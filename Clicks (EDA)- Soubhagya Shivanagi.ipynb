# Univariate Analysis

### Categorical Columns

```python
df2.select_dtypes(include=object).columns
```

```python
plt.rcParams['figure.figsize']=[15,8]
```

```python
cols = ['Created_On', 'Source', 'Industry']
rows = 3
columns = 1
iterator = 1

for i in cols:
    plt.subplot(rows, columns, iterator)
    
    if i == 'Created_On':
        # Use a bar plot for datetime data
        sns.countplot(data=df2, x=i)
    else:
        sns.countplot(data=df2, y=i)
    
    plt.title(i)
    iterator += 1
    plt.xticks(rotation=90)

plt.tight_layout()
plt.show()
```
