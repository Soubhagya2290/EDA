# Univariate Analysis

### Categorical Columns

```python
df5.select_dtypes(include=object).columns
```

```python
plt.rcParams['figure.figsize']=[15,8]
```

```python
cols = ['Sales_Agent', 'Manager', 'Regional_Office']
rows = 3
columns = 1
iterator = 1

for i in cols:
    plt.subplot(rows, columns, iterator)
    sns.countplot(data=df5, x=i)  # Use 'data=df5' to specify the DataFrame
    plt.title(i)
    iterator += 1
    plt.xticks(rotation=90)

plt.tight_layout()
plt.show()
```

# Bivariate Analysis

#### Cat vs Cat

```python
cols = ['Sales_Agent', 'Manager', 'Regional_Office']
rows = 3
columns = 1  # Set to 1 since you want each column to have its own subplot
iterator = 1

plt.figure(figsize=(10, 15))  # Adjust the figure size as needed

for i in cols:
    plt.subplot(rows, columns, iterator)
    sns.countplot(data=df5, x=i, hue='Manager')  # Assuming 'Manager' is the column you want to use as hue
    plt.title(i)
    iterator += 1
    plt.xticks(rotation=90)

plt.tight_layout()
plt.show()
```
